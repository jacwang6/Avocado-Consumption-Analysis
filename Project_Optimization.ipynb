### Organic Avocados Optimization

```python
ampl1 = AMPL(Environment('C:\\Users\\sheds\\Desktop\\ampl'))
ampl1.setOption('solver', 'CPLEX')
ampl1.read('Project_Organic.mod')
ampl1.readData('Project_Organic.dat')
ampl1.solve()
```

```python
x = ampl1.getVariable('x')
for index, i in x:
    if (i.value()>0.0): print(index, i.value())
```
